```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Sales_Data.csv', encoding= 'unicode_escape') 
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Zipcode', axis=1, inplace=True) 
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Gender'] = df['Gender'].replace('M', 'Male') 
```

```python
# TOTAL TRANSACTION COUNT BY GENDER IN BAR CHART 
sns.countplot(x='Gender' , data=df)
plt.show() 

```

```python
ax =sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    ax.set_title('Gender Distribution', fontsize=16)  #chart title
    ax.set_xlabel('Gender', fontsize=16)              #X-Axis 
    ax.set_ylabel('Count', fontsize=16)               #Y-Axis
    ax.tick_params(axis='both', labelsize=10)         #Axis Parameters
plt.show() 
```

```python

gender_counts = df["Gender"].value_counts() 
print(gender_counts) 
```

```python
#TOTAL TRANSACTION COUNT BY GENDER WISE
gender_count = df['Gender'].value_counts() 

plt.pie( 
    gender_counts,                #Data (counts)
    labels=gender_counts.index,   #labels(male,female)
    autopct= '%1.2f%%',           #show percentages (1 decimle place)
    startangle=90,                #rotate for better orientation
    
)
#ADD TITLE
plt.title('Gender Distribution', fontsize=16)
plt.show() 
```

```python
df.groupby('Gender', as_index=False)[['Amount', 'Orders']].sum()  
```

```python
df.groupby('Gender')['Amount'].sum() 
```

```python
#GENDER WISE TOTAL SALES

#STEP 1; GROUP DATA BY GENDER AND SUM() AMOUNT
gen_wise_total = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

#STEP 2: CREATE THE BAR PLOT 
ax = sns.barplot(x='Gender' , y='Amount', data=gen_wise_total) 

# STEP 3: ADD DATA LABELS ON TOP OF THE BAR

for container in ax.containers:
    ax.bar_label(ax.containers[0], fmt="%.0f", fontsize=12, color="black")
    
#STEP 4: ADDING TITLE 
plt.title('Gender_wise_Toal_Sales', fontsize=14)
plt.show() 
```

### AGE GROUP WISE TOTAL TRANSACTIONS

```python
age_count=df['Age Group'].value_counts().sort_values(ascending=True)
sns_orders=age_count.index
ax = sns.countplot(data=df, x='Age Group', order=sns_orders)

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

```python
#AGE GROUP AND GENDER WISE TRANSACTION DISTRIBUTION

ax = sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

```python
#AGE GROUP WISE TOTAL AMOUNT
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age) 
plt.show() 
```

### STATE WISE ANALYSIS

```python
#ORDERS WISE TOP 5 STATES

order_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5) 

ax=sns.barplot(data=order_state,x='State',y='Orders')

for bars in ax.containers:
    ax.bar_label(bars) 
plt.show() 
```

```python
#AMOUNT WISE TOP 5 STATES
order_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)}) 
sns.barplot(data=order_state,x='State',y='Amount')
plt.show()
```

### PRODUCT CATEGORY ANALYSIS 

```python
#PRODUCT CATEGORY WISE TRANSACTION COUNT 
ax = sns.countplot(data=df, x='Product_Category')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars) 
plt.show() 
```

```python
#AMOUNT WISE PRODUCT CATEGORY IN BAR CHART 
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount') 
plt.show() 
```

```python
#PRODUCT CATEGORY AND GENDER WISE TOTAL COUNT 
sales_pro=df.groupby(['Product_Category', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_pro,x='Product_Category',y='Amount', hue ='Gender') 
plt.show() 
```

### PROFESSION WISE ANALYSIS 

```python
#AMOUNT WISE TOP PROFESSION
ax = sns.countplot(data = df, x='Profession')
sns.set(rc={'figure.figsize':(5,5)}) 
for bars in ax.containers:
    ax.bar_label(bars) 
plt.show() 
```

```python
#AMOUNT WISE PROFESSION 
sales_pro=df.groupby(['Profession'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_pro,x='Profession',y='Amount') 
plt.show() 
```

```python
#ORDER WISE TOP 10 PRODUCTS 
sales_state=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_ID',y='Orders') 
plt.show() 
```
